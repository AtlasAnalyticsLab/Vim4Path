```python
import os
import pandas as pd
from natsort import os_sorted 
```

```python
experiments = os_sorted(os.listdir('results/'))
aucs  = []
accs = []
f1s = []
valid_experiments= []
for experiment in experiments:
    if os.path.exists(os.path.join(f'results/{experiment}/summary.csv')):
        df = pd.read_csv(os.path.join(f'results/{experiment}/summary.csv'))
        aucs.append(df['test_auc'].iloc[0])
        accs.append(df['test_acc'].iloc[0])
        f1s.append(df['test_f1'].iloc[0])
        valid_experiments.append(experiment)
```

```python
df = pd.DataFrame([valid_experiments, aucs, accs, f1s]).T
df.columns = ['Experiment', 'Auc', 'Acc', 'F1']
df.sort_values('F1')
```

```python
df.to_csv('mil.csv')
```
